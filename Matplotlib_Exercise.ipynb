```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x *2
z = x**2
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('Diagram')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title1')

axes2 = fig.add_axes([0.2,0.5,.2,.2])
axes2.plot(x,y,'r')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('title2')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)



axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2 = fig.add_axes([0.2,0.6,.2,.2])
axes2.plot(x,y)
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('zoom')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,lw=3,ls='--')
axes[1].plot(x,y,lw=3,ls='-')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color = 'r',lw=3,ls='--')
axes[1].plot(x,y,color = 'r',lw=3,ls='-')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize =(12,2))
axes[0].plot(x,y,color = 'r',lw=3,ls='--')
axes[1].plot(x,y,color = 'r',lw=3,ls='-')
```
